```python
df=apoe.merge(phs, on='RID', how='inner')
df[['allele1', 'allele2']] = df['APOE Genotype'].str.split(",", expand = True)
```

```python
df.columns
```

```python
columns=df[['AGE','PTGENDER','PTEDUCAT','allele1','allele2','PTEDUCAT','APOE4','MMSE', 'PHS','CIR','DX.bl']]
```

```python
from sklearn.utils import resample
df3=pd.DataFrame(columns)
df3
class_counts = df["DX.bl"].value_counts()
classes = class_counts.index.tolist()
```

```python
resampled_dfs = []
```

```python
for class_ in classes:
    class_df = df3[df3["DX.bl"] == class_]
    majority_class_df = resample(class_df, n_samples=class_counts.max(), random_state=42)
    resampled_dfs.append(majority_class_df)
```

```python
resampled_df = pd.concat(resampled_dfs)
resampled_df['DX.bl'].value_counts()
df3 = resampled_df.sample(frac=1, random_state=42).reset_index(drop=True)
df3
```

```python
columns=df3[['AGE','PTGENDER','PTEDUCAT','allele1','allele2','PTEDUCAT','APOE4','MMSE', 'PHS','CIR','DX.bl']]
```
