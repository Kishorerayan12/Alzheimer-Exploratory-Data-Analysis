```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/NEUROPATH_11_14_22 of alzheimer patients  updated.csv')
df
```

```python
for col in df.columns:
    unique_values = df[col].unique()
    print(f'Unique values in {col}: {unique_values}')
```

```python
unique_counts = df.nunique()

# Select columns with more than one unique value
cols_to_keep = unique_counts[unique_counts > 1].index

# Drop columns with only one unique value
df = df[cols_to_keep]

print(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr=df.corr()
fig, ax = plt.subplots(figsize=(20, 18))
sns.heatmap(corr, cmap='coolwarm', annot=True, ax=ax)
plt.show()
```

```python
for col in df.columns:
    unique_values = df[col].unique()
    print(f'Unique values in {col}: {unique_values}')
```

```python
for col in df.columns[3:9]:
  df[col]=df[col].replace([0,1,2,3,8,9],['None', 'Mild', 'Mderate', 'Severe', 'Not assesed','unknown'])

for col in df.columns[10:11]:
  df[col]=df[col].replace([1,2,8],['non phospho specific','Phospho specific','missing'])

df['NPNEUR']=df['NPNEUR'].replace([0,1,2,3],['No neuritic plaques','Sparse neuritic plaques', ' Moderate neuritic plaques','Frequent neuritic plaques'])
df['NPADNC']=df['NPADNC'].replace([0,1,2,3],['no AD', 'low', 'intermediate','high'])
df['NPDIFF']=df['NPDIFF'].replace([0,1,2,3],['No neuritic plaques','Sparse neuritic plaques', ' Moderate neuritic plaques','Frequent neuritic plaques'])
df['NPAMY']=df['NPAMY'].replace([0,1,2,3],['no', 'low', 'intermediate','high'])
df['NPINF']=df['NPINF'].replace([0,1,9],['no','yes','missing'])
df['NPHEMO']=df['NPHEMO'].replace([0,1,9],['no','yes','missing'])
df['NPOLD']=df['NPOLD'].replace([0,1,9],['no','yes','missing'])
df['NPARTER']=df['NPARTER'].replace([0,1,2,3],['no', 'mild', 'moderate','severe'])
df['NPWMR']=df['NPWMR'].replace([0,1,2,3],['no', 'mild', 'moderate','severe'])
df['NPPATH']=df['NPPATH'].replace([0,1,9],['no','yes','missing'])
df['NPLBOD']=df['NPLBOD'].replace([0,1,2,3,4,5],['No','Brainstem predominant','Limbic (transitional)','Neocortical (diffuse)', 'Amygdala predominant', 'Olfactory bulb'])
df['NPPDXI']=df['NPPDXI'].replace([0,1,9],['no','yes','missing'])
df['NPPDXK']=df['NPPDXK'].replace([0,1,9],['no','yes','missing'])
df['NPPDXM']=df['NPPDXM'].replace([0,1,9],['no','yes','missing'])
df['NPPDXP']=df['NPPDXP'].replace([0,1,9,8],['no','yes','missing','not assesed'])
df['NPPDXQ']=df['NPPDXQ'].replace([0,1,9,8],['no','yes','missing','not assesed'])
```

```python
import seaborn as sns
fig, ax = plt.subplots()
sns.boxplot(x='NPADNC', y='NPWBRWT', data=df, ax=ax)

# set the y-axis limit and tick labels
ax.set_ylim(600, 2000)
ax.set_yticks(np.arange(0, 2500, 100))

# show the plot
plt.show()
```

```python
dependent_column = 'NPADNC'
columns_to_plot = ['NPGRCCA', 'NPGRLA', 'NPGRHA', 'NPGRSNH',
       'NPGRLCH', 'NPAVAS', 'NPTAN', 'NPTDPAN', 'NPBRAAK', 'NPNEUR',
       'NPDIFF', 'NPAMY', 'NPINF', 'NPHEMO', 'NPOLD', 'NPARTER', 'NPWMR',
       'NPPATH', 'NPLBOD', 'NPPDXI', 'NPPDXK', 'NPPDXM', 'NPPDXP', 'NPPDXQ']  # list of columns to plot
for col in columns_to_plot:
    plt.figure(figsize=(6, 6))
    sns.countplot(x=col, hue=dependent_column, data=df)
    plt.title(f"Count plot for {col} against {dependent_column}")
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```
