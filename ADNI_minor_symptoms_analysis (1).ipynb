```python
ref_var='DX.bl'
cols_plot= ['AXNAUSEA', 'AXVOMIT', 'AXDIARRH', 'AXCONSTP', 'AXABDOMN',
       'AXSWEATN', 'AXDIZZY', 'AXENERGY', 'AXDROWSY', 'AXVISION', 'AXHDACHE',
       'AXDRYMTH', 'AXBREATH', 'AXCOUGH', 'AXPALPIT', 'AXCHEST', 'AXURNDIS',
       'AXURNFRQ', 'AXANKLE', 'AXMUSCLE', 'AXRASH', 'AXINSOMN', 'AXDPMOOD',
       'AXCRYING', 'AXELMOOD', 'AXWANDER', 'AXFALL', 'AXOTHER']
```

```python
for col in cols_plot:
    # filter data by reference variable
    group_data = df[df['DX.bl'] == "AD"]
    if "yes" in group_data[col].value_counts():
      group_count = group_data[col].value_counts()["yes"]
      group_avgAD= (group_count / len(group_data))*100
      print(f"Average count of yes for AD in {col} column:", group_avgAD)
    else:
      print(f"No 'yes' values found in {col} column")
    
    group_data = df[df['DX.bl'] == "LMCI"]
    if "yes" in group_data[col].value_counts():
      group_count = group_data[col].value_counts()["yes"]
      group_avgLMCI = (group_count / len(group_data))*100
      print(f"Average count of yes for LMCI in {col} column:", group_avgLMCI)
    else:
      print(f"No 'yes' values found in {col} column")

    group_data = df[df['DX.bl'] == "EMCI"]
    if "yes" in group_data[col].value_counts():
      group_count = group_data[col].value_counts()["yes"]
      group_avgEMCI = (group_count / len(group_data))*100
      print(f"Average count of yes for EMCI in {col} column:", group_avgEMCI)
    else:
      print(f"No 'yes' values found in {col} column")
    
    group_data = df[df['DX.bl'] == "CN"]
    if "yes" in group_data[col].value_counts():
      group_count = group_data[col].value_counts()["yes"]
      group_avgCN = (group_count / len(group_data))*100
      print(f"Average count of yes for CN in {col} column:", group_avgCN)
    else:
      print(f"No 'yes' values found in {col} column")
    fig, ax = plt.subplots()
    ax.bar(["AD", "LMCI",'EMCI', 'CN'], [group_avgAD, group_avgLMCI, group_avgEMCI, group_avgCN ])
    ax.set_xlabel(ref_var)
    ax.set_ylabel(f"Average count of yes for {col} column")
    ax.set_title(f"{col} vs {ref_var}")
    plt.show()
```
